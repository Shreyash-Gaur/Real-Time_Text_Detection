```python
# Initialize the webcam
cap = cv2.VideoCapture(0)  # Adjust the index if needed

# Instance text detector
reader = easyocr.Reader(['en'], gpu=True)

threshold = 0.25

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    if not ret: 
        print("Error: Failed to capture frame from webcam")
        continue 
        
    # Detect text on the frame
    text_ = reader.readtext(frame)

    # Draw bbox and text on the frame
    for t_, t in enumerate(text_):
        bbox, text, score = t

        if score > threshold:
            print(bbox)  # Print the bbox for inspection

            # Assuming bbox is a list of 4 points, adjust if needed
            pt1 = (int(bbox[0][0]), int(bbox[0][1]))
            pt2 = (int(bbox[2][0]), int(bbox[2][1]))

            cv2.rectangle(frame, pt1, pt2, (0, 255, 0), 5)
            cv2.putText(frame, text, pt1, cv2.FONT_HERSHEY_COMPLEX, 0.65, (255, 0, 0), 2)

    # Display the resulting frame
    cv2.imshow('Webcam Text Detection', frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture and close all windows
cap.release()
cv2.destroyAllWindows()
```
